```python
import pandas as pd
```

```python
df = pd.read_csv('data/training_set-new_set.csv')
df.head()
```

```python
from scipy.fftpack import fft
import numpy as np

close_fft = fft(np.asarray(df['close'].tolist()))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7), dpi=100)
plt.plot(close_fft, color='blue')
plt.title('Fourier transform apple stock')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.show()
```
